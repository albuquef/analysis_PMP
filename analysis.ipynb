```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
PATH_DATA='tables/'

df_mat_p_26_arrond = pd.read_csv(PATH_DATA+'reports_mat_p_26_cover_arrond.csv')
df_mat_p_42_arrond = pd.read_csv(PATH_DATA+'reports_mat_p_42_cover_arrond.csv')
df_mat_p_51_epci = pd.read_csv(PATH_DATA+'reports_mat_p_51_cover_epci.csv')
df_mat_p_62_epci = pd.read_csv(PATH_DATA+'reports_mat_p_62_cover_epci.csv')

df_urgenc_p_42_arrond = pd.read_csv(PATH_DATA+'reports_urgenc_p_42_cover_arrond.csv')
df_urgenc_p_66_arrond = pd.read_csv(PATH_DATA+'reports_urgenc_p_66_cover_arrond.csv')
df_urgenc_p_54_epci = pd.read_csv(PATH_DATA+'reports_urgenc_p_54_cover_epci.csv')
df_urgenc_p_78_epci = pd.read_csv(PATH_DATA+'reports_urgenc_p_78_cover_epci.csv')
```

```python

df = df_mat_p_26_arrond
p = 26
cover = 'arrond'
# df.plot(x='time_exact', y='sol_exact', kind='line')
# plt.plot(df.exact_time, df.exact_sol, color='red', label='Exact',linestyle="--")
plt.plot(df.exact_TL_time, df.exact_TL_sol, color='red', label='Exact')
plt.plot(df.rssv_tbPMP_exact_time, df.rssv_tbPMP_exact_sol, color='orange', label='RSSV-exact')
plt.plot(df.rssv_tbPMP_exact_vns_time, df.rssv_tbPMP_exact_vns_sol, color='green', label='RSSV-exact-vns')



# Add labels and legend
plt.xlabel('Real Time')
plt.ylabel('FO')
plt.legend()
plt.title(f'Maternity p = {p} cover {cover} (18)')

# Save the plot as a PDF file
plt.savefig(f'maternity_plot_p_{p}_{cover}.pdf')

# Show the plot (optional)
plt.show()
```
